```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel("responses.xlsx")
data.head(10)
```

```python
data.columns
```

```python
data.columns=['n','datosp','correo','edad','genero','barrio','analizar1','op1','op2','op3','op4','op5','op6','op7','op8','op9','op10','analizar2','fecha','entrada','iduser']
```

```python
data.head()
```

```python
dataanaliz1 = (data['analizar1'].tolist())
print(dataanaliz1)
```

```python
strlist = " ".join(map(str, dataanaliz1))
print(strlist)
type(strlist)
```

```python
nandel = strlist.replace("nan","")
print(nandel)
```

```python
verifi = nandel.find("nan")
print(verifi)
```

```python
allminus = nandel.lower()
```

```python
tokens = [palabra for palabra in allminus.split()]
print (tokens)
type (tokens)
```

```python
a = stopwords.words('spanish')
print(a)
```

```python
tokens_limpios = tokens[:]
for token in tokens:
    if token in a:
        tokens_limpios.remove(token)
```

```python
frequencia = nltk.FreqDist(tokens_limpios)
for palabra,valor in frequencia.items():
    print(str(palabra)+ ':' +str(valor))
```

```python
plt.figure(figsize=(15, 5))
frequencia.plot(30, cumulative=False)
```
